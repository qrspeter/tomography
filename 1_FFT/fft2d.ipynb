````python
import numpy as np
import matplotlib.pyplot as plt

from skimage.data import shepp_logan_phantom, brick
from skimage.transform import radon, rescale, iradon

image = brick()
#image = shepp_logan_phantom()
#image = rescale(image, scale=0.4, mode='reflect', channel_axis=None)
````

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8))

ax1.set_title("Original")
ax1.imshow(image, cmap=plt.cm.Greys_r)

image_fft2 = np.fft.fft2(image)
ax2.set_title("FFT")
ax2.imshow(abs(image_fft2), cmap=plt.cm.Greys_r)

ax3.set_title("Log FFT")
ax3.imshow(np.log(abs(image_fft2)), cmap=plt.cm.Greys_r)

ax4.set_title("Log FFT shifted")
ax4.imshow(np.log(abs(np.fft.fftshift(np.fft.fft2(image)))), cmap=plt.cm.Greys_r)

fig.tight_layout()
plt.show()
````

### Двумерное Фурье-преобразование в GNU Octave
Фурье-преобразование можно визуализировать так (к абсолютным значениям добавлена 1 для лучшего отображения):
```MATLAB
pkg load image;
P = phantom(256);

title('FFT');

imshow(log(abs(F)+1),[])
title('log - FFT'); 

F = fftshift(F); # rearranges a Fourier transform X by shifting the zero-frequency component to the center of the array
subplot(2,2,4);
imshow(log(abs(F)+1),[])
title('centered - log - FFT');
```
